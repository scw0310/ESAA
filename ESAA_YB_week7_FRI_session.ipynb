3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
train['type'] = train['type'].map({'white': 0, 'red': 1})
test['type'] = test['type'].map({'white': 0, 'red': 1})
```

```python
train.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.

train_x = train.drop(columns=['index', 'quality'])
train_y = train['quality']

test_x = test.drop(columns=['index'])
```

```python
train_x.head()
```

```python
train_y.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error

train_data = lgb.Dataset(train_x, label=train_y)

model = lgb.train({}, train_data, num_boost_round=1000)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
y_test_pred = model.predict(test_x, num_iteration=model.best_iteration)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality'] = y_test_pred
```

```python
smpl_sub.head()
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

train_one = pd.get_dummies(data = train, columns = ['type'], prefix = ['type'])
test_one = pd.get_dummies(data = test, columns = ['type'], prefix = ['type'])
encoder = LabelEncoder()
train_one['quality'] = encoder.fit_transform(train_one['quality'])

train_one.head()
```

```python
test_one.head()
```

```python
# 모델 정의
import xgboost as xgb
from sklearn.metrics import accuracy_score

# 'quality' 열을 목표 변수
X_train = train_one.drop(columns=['quality'])
y_train = train_one['quality']

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수


# fit 메소드를 이용해 모델 학습
xgb_model.fit(X_train, y_train)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
y_test_pred = xgb_model.predict(test_one)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
submission = pd.read_csv('/content/drive/MyDrive/EWHA/ESAA/data/wine_data/sample_submission.csv')

submission['quality'] = y_test_pred
submission.head()
```

```python
# 정답파일 내보내기
submission.to_csv('xgb_pred.csv',index = False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train = pd.get_dummies(train)
test = pd.get_dummies(test)

test = test.reindex(columns=train.columns, fill_value=0)

train.head()
```

```python
test.head()
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
model1 = LGBMClassifier()
model2 = XGBClassifier()
model3 = RandomForestClassifier()

# VotingClassifier 정의
voting_clf = VotingClassifier(estimators=[
    ('lgbm', model1),
    ('xgb', model2),
    ('rf', model3)
], voting='hard')
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
# fit 메소드를 이용해 모델 학습
X = train.drop(columns=['quality'])
y = train['quality']
voting_clf.fit(X, y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
voting_clf.predict(test.drop(columns=['quality']))
```
