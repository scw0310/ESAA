````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

DataFrame1 = pd.DataFrame(raw_data)
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(10,6))
plt.scatter(DataFrame1['preTestScore'], DataFrame1['postTestScore'],
            s=DataFrame1['age']*10, color='blue')

plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Scatter Plot of Pre-Test vs Post-Test Scores (Size by Age)')
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
DataFrame1['female']
````

````python
plt.figure(figsize=(10,6))
plt.scatter(DataFrame1['preTestScore'], DataFrame1['postTestScore'],
            s=DataFrame1['postTestScore']*4.5,
            c=['pink' if female == 1 else 'skyblue' for female in DataFrame1['female']])

plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Scatter Plot of Pre-Test vs Post-Test Scores (Size by Post-Test Score, Color by Gender)')

plt.scatter([], [], color='skyblue', label='Male')
plt.scatter([], [], color='pink', label='Female')
plt.legend(title="Gender")
````
