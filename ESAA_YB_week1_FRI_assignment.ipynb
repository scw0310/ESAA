# **개념정리**

### **DataFrame 시작**


*   .shape() : DataFrame의 행과 열을 튜플 형태(행, 열)로 반환
*   .info() : 총 데이터 건수, 데이터 타입, Null건수
*   .describe() : n-pertile 분포도, 평균, 최대, 최소. 숫자형 칼럼의 분포도만 조사
*   DataFrame[].value_counts() : 해당 칼럼값 유형, 건수 확인. value_counts(dropna=False)는 Null값 포함하여 계산





### **DataFrame 데이터 삭제**


*   **.drop(labels=None, axis=0, index=None, columns=None, inplace=False, errors'raise')**
*   axis : 드롭 방향. 0은 로우 축, 1은 칼럼 축
*   labels : 원하는 칼럼명



### **데이터 셀렉션 및 필터링**


*   DataFrame[col] : 혼돈 방지 위해 칼럼만 지정하는 것으로 사용
*   DataFrame[DataFrame['colname'] == 3] : 불린 인덱싱



### **iloc, loc 연산자**


*   iloc[행, 열] : 위치 기반, 칼럼 명 입력시 오류 발생. iloc[:, -1]은 맨 마지막 칼럼 데이터, iloc[:, :-1]은 맨 마지막 칼럼 제외 모든 데이터
*   loc[인덱스 값, 칼럼 명] : 명칭 기반

**불린 인덱싱**
*   and : &
*   or : |
*   not : ~
예시) titanic[(titanic['Age'] > 60) & titanic['Pclass'] == 1]

### **정렬, Aggregation 함수, GroupBy 적용**
**sort_values()**
*   ascending = True (오름차순), ascending = False (내림차순)
*   inplace = False : 원 DataFrame 유지, 정렬한 DataFrame 반환

**groupby(by = col)**
: 칼럼 기준으로 groupby하여 반환

### **결손 데이터 처리**


*   isna() : NaN 여부 확인
*   fillna() : NaN 다른 값으로 대체

### **apply lambda 식으로 데이터 가공**

```python
def get_square(a):
  return a**2
print('3의 제곱은:', get_square(3))
```

```python
lambda_square = lambda x : x ** 2
print('3의 제곱은:', lambda_square(3))
```

```python
a=[1, 2, 3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
def get_category(age) :
  cat = ''
  if age <= 5 : cat = 'Baby'
  elif age <= 12 : cat = 'Child'
  elif age <= 18 : cat = 'Teenager'
  elif age <= 25 : cat = 'Student'
  elif age <= 35 : cat = 'Young Adult'
  elif age <= 60 : cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
